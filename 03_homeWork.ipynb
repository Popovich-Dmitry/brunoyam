```python
Titanic = pd.read_csv('C:/Users/wwwpo/Documents/titanic.csv')
```

```python
Total = len(Titanic) #Общее количество 891
print(Total, 'пассажиров на корабле')
```

```python
#Age_defined = len(Titanic.loc[Titanic['Age'].isnull()]) #Возраст определен 714
```

```python
Kids = len(Titanic[Titanic['Age']<18]) #Дети 139
print(Kids, 'Детей на корабле')
```

```python
Kids_from_Total= Kids*100/Total #Процент детей 0.156%
print(round(Kids_from_Total,2), 'Процент детей на корабле')
```

```python
Female = len(Titanic.loc[Titanic['Sex']== 'female'])*100/Total #Женщины 0.35%
print(round(Female,2), 'Процент женщин на корабле')
```

```python
Male = len(Titanic.loc[Titanic['Sex']== 'male'])*100/Total #Мужчины 0.65%
print(round(Male,2), 'Процент мужчин на корабле')
```

```python
men_survived = len(Titanic.loc[(Titanic['Sex']=='male') & (Titanic['Survived'].astype(int)==1)])*100/\
                    len(Titanic.loc[Titanic['Sex']=='male'])
print(round(men_survived,2), 'Мужчин выжило')
women_survived = len(Titanic.loc[(Titanic['Sex']=='female') & (Titanic['Survived'].astype(int)==1)])*100/\
                    len(Titanic.loc[Titanic['Sex']=='female'])
print(round(women_survived,2), 'Женщин выжило')
#Процент выживших детей
kids_survived = len(Titanic.loc[(Titanic['Age']<18) & (Titanic['Survived'].astype(int)==1)])*100/\
                    len(Titanic.loc[Titanic['Age']<18])
print(round(kids_survived,2), 'Детей выжило')
```
