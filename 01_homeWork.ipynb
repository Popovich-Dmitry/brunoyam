# Easy
Нужно написать функцию, которая на вход получает список из чисел, а на выходе дает список, где каждое число возведено в квадрат. Решать при помощи цикла.

Попробовать на вход:

# 1.
list(range(10))

# 2.
list(range(100))

# 3.
list()

```python
#1
def Easy (list_1):
    new_list =  []
    for i in list_1:
        new_list.append(i*i)
    return new_list

list_1 = [x for x in range(10)]
Easy(list_1)
```

```python
#2
list_2 = [x for x in range(100)]
Easy(list_2)
```

```python
#3

list_3 = ()
Easy(list_3)
```

# Hard
Написать функцию, которая на вход принимает список, а возвращает словарь со среднием, медианой и модой, например вот такой: {"mean": ..., "median": ..., "mode": ...}. На следующем занятии разберем как это сделать при помощи библиотек, а пока нужно реализовать просто по формулам.
Пример:

func([0, 1, 1, 10, 5, 4, 3])

```python
def hard (list):
    mean = sum(list)/len(list)
    list = sorted(list)
    if len(list) % 2 ==0:
        median = (list[len(list)//2-1]+list[len(list)//2])/2
    else:
        median = list[len(list)//2]
        
    counts = []
    for i in list:
        counts.append(list.count(i))
    maximum = max(counts)
    index = counts.index(maximum)
    moda = list[index]
    return {'mean':mean, 'median':median, 'moda':moda}
print(hard([0, 1, 1, 10, 5, 4]))
```
